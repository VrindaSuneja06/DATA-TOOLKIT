## *Theory : *

1.What is NumPy, and why is it widely used in Python?
NumPy is a library for numerical computing in Python. It is used because it provides fast and efficient operations on large arrays and matrices.

2.How does broadcasting work in NumPy?
Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array.

3.What is a Pandas DataFrame?
A DataFrame is a table-like data structure in Pandas that stores data in rows and columns, similar to a spreadsheet.

4.Explain the use of the groupby() method in Pandas.
The groupby() method groups data based on a column and allows applying functions like sum, mean, or count to each group.

5.Why is Seaborn preferred for statistical visualizations?
Seaborn provides beautiful and easy-to-use charts for statistical data analysis. It includes built-in themes and complex visualizations.

6.What are the differences between NumPy arrays and Python lists?
NumPy arrays are faster, use less memory, and support mathematical operations directly, while Python lists are slower and require loops for calculations.

7.What is a heatmap, and when should it be used?
A heatmap is a color-coded grid that shows relationships between values in a matrix. It is used for correlation analysis or data comparison.

8.What does the term “vectorized operation” mean in NumPy?
It means performing operations on entire arrays without using loops, making computations faster.

9.How does Matplotlib differ from Plotly?
Matplotlib is static and good for basic plots, while Plotly is interactive and better for complex visualizations.

10.What is the significance of hierarchical indexing in Pandas?
It allows multiple levels of indexing in DataFrames, making it easier to organize and access complex data.

11.What is the role of Seaborn’s pairplot() function?
It creates scatter plots for every pair of numerical columns in a dataset, helping to explore relationships.

12.What is the purpose of the describe() function in Pandas?
It gives summary statistics like mean, count, and standard deviation of numerical columns in a DataFrame.

13.Why is handling missing data important in Pandas?
Missing data can affect analysis accuracy. Pandas provides ways to fill or remove missing values.

14.What are the benefits of using Plotly for data visualization?
Plotly supports interactive charts, zooming, and 3D visualizations, making it better for presentations and dashboards.

15.How does NumPy handle multidimensional arrays?
NumPy stores arrays in multiple dimensions (1D, 2D, 3D, etc.), allowing efficient mathematical operations.

16.What is the role of Bokeh in data visualization?
Bokeh is used for interactive visualizations and is suitable for web applications.

17.Explain the difference between apply() and map() in Pandas.
apply() is used on DataFrames to apply functions to rows or columns, while map() works only on Series (single columns).

18.What are some advanced features of NumPy?
Linear algebra functions, Fourier transforms, random number generation, and broadcasting.

19.How does Pandas simplify time series analysis?
Pandas provides built-in support for date-time indexing, resampling, and time-based calculations.

20.What is the role of a pivot table in Pandas?
A pivot table summarizes large data by grouping and applying functions like sum or mean.

21.Why is NumPy’s array slicing faster than Python’s list slicing?
NumPy arrays use continuous memory blocks, while Python lists store elements separately, making slicing slower.

22.What are some common use cases for Seaborn?
Visualizing correlations, distributions, categorical data, and statistical relationships.

```python
#3
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

plt.scatter(x, y, color='blue', marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

```python
#4
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = np.random.rand(5, 5)
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#5
import plotly.express as px

data = {'Category': ['A', 'B', 'C', 'D'], 'Values': [10, 20, 15, 25]}
df = pd.DataFrame(data)

fig = px.bar(df, x='Category', y='Values', title="Bar Plot Example")
fig.show()
```

```python
#8
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label="Line 1", color='r')
plt.plot(x, y2, label="Line 2", color='b')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.title("Multiple Line Plot")
plt.show()
```
